## Az előzőek ismeretében már le tudjuk tölteni a tartalmat:

```python
import pandas as pd
from newspaper import Article

# Adatok betöltése DataFrame-be
df = pd.read_csv("telex_english_news.csv")

# Cikkek tartalmának letöltése
for index, row in df.iterrows():
    link = row['Link']
    article = Article(link)
    article.download()
    article.parse()
    text = article.text
    # A cikk tartalmának hozzáadása a DataFrame-hez
    df.at[index, 'Text'] = text

df.to_csv("telex_news_with_summary.csv", index=False)

# DataFrame tartalmának kiírása
print(df)
print(len(df))
#print(df.loc[9, 'Text'])
```

# Elkészítjük a modelt.

```python
import pandas as pd
from newspaper import Article
from transformers import BartTokenizer, TFBartForConditionalGeneration
from transformers import pipeline
from googletrans import Translator

# Adatok betöltése DataFrame-be
df = pd.read_csv("telex_news_with_summary.csv")

# BART model betöltése
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = TFBartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

# Érzelmi analízis létrehozása
sentiment_classifier = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest")

# Fordítás angolról magyarra
translator = Translator()

# Összefoglalás létrehozása
def generate_summary(text):
    if isinstance(text, str):
        text = [text]  # Ha csak egy stringet kapunk, alakítsuk át listává
    inputs = tokenizer(text, max_length=1024, return_tensors='tf', truncation=True, padding=True)
    summary_ids = model.generate(inputs['input_ids'], max_length=80, num_beams=4, early_stopping=True)
    summaries = [tokenizer.decode(summary_id, skip_special_tokens=True) for summary_id in summary_ids]
    return summaries[0] if len(summaries) == 1 else summaries  # Ha több összefoglalást generálunk, csak az elsőt adjuk vissza

# Érzelmi hangulat értékelése
def evaluate_sentiment(summary):
    result = sentiment_classifier(summary)
    return result[0]

# Fordítás angolról magyarra
def translate_to_hungarian(summary):
    translation = translator.translate(summary, src='en', dest='hu')
    return translation.text

# Összefoglalások, érzelmi hangulat és fordítás hozzáadása a DataFrame-hez
df['Summarized'] = df['Text'].apply(generate_summary)
df['Sentiment'] = df['Summarized'].apply(evaluate_sentiment)
df['Translated_Summary'] = df['Summarized'].apply(translate_to_hungarian)

# Emailhez előkészített DataFrame létrehozása HTML formátumban
email_df = df[['Title', 'Link', 'Summarized', 'Sentiment', 'Translated_Summary']]
email_html = email_df.to_html(index=False)
```
